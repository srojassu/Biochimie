````python
#Definir los vectores
X=[1, 2, 3, 4, 5]
Y=[10, 11, 30, 31, 32]
````

Correr el siguiente bloque para realizar el ajuste

````python
#Se definen parametros para el bucle
funciones=list(Diccionario_funciones.keys())
j=0
Numerofunciones=len(funciones)

#Se definen vectores que guardaran los resultados de cada función en el bucle
Primer_Parametro=np.ones((Numerofunciones,1))
Segundo_Parametro=np.ones((Numerofunciones,1))
Tercer_Parametro=np.ones((Numerofunciones,1))
Cuarto_Parametro=np.ones((Numerofunciones,1))
Sum_difcuadrada=np.ones((Numerofunciones,1)) #vector guardar los valores de la sumatoria de los errores cuadrados
ERROR=np.zeros((Numerofunciones,1)) #Vector por si no se encuentran parametros que ajustan, 0 indica que esa función no tiene error
Matriz_resultados=np.zeros((Numerofunciones,2))  #Matriz para guardar resultados
#Se inicia el bucle
while j<Numerofunciones:
  error=False
  try:
    parametros, covarianza = curve_fit(Diccionario_funciones[j],X,Y) #Ajusta los parametros a los datos
  except:
    error=True
  #guardamos los parametros en los vectores definidos
  if error is False:
    Primer_Parametro[j]=parametros[0]
    Segundo_Parametro[j]=parametros[1]
    Tercer_Parametro[j]=parametros[2]
    Cuarto_Parametro[j]=parametros[3]
    Funcion_evaluada=Diccionario_funciones[j](X,Primer_Parametro[j],Segundo_Parametro[j], Tercer_Parametro[j], Cuarto_Parametro[j])
    calculo_dif=np.array(Y-Funcion_evaluada)
    Calculo_difcuadrada=calculo_dif**2
    Calculo_sum_difcuadrada=sum(Calculo_difcuadrada)
    Sum_difcuadrada[j]=Calculo_sum_difcuadrada
    Matriz_resultados[j,0]=j+1
    Matriz_resultados[j,1]=Sum_difcuadrada[j]
  else:
    Primer_Parametro[j]=0
    Segundo_Parametro[j]=0
    Tercer_Parametro[j]=0
    Cuarto_Parametro[j]=0
    Sum_difcuadrada[j]=mt.inf
    ERROR[j]=1
    Matriz_resultados[j,0]=j+1
    Matriz_resultados[j,1]=Sum_difcuadrada[j]
  j=j+1
Matriz_ordenada=Matriz_resultados[Matriz_resultados[:,1].argsort()]
print("A continuación se presentan ordenadas las funciones de menor a mayor variación o desviación")
print(["Función","desviación"])
print(Matriz_ordenada)
#Definimos funciones para visualizar los parametros obtenidos
#(Primer_Parametro)
#print(Segundo_Parametro)
#print(Tercer_Parametro)
#print(Cuarto_Parametro)
#print(Quinto_Parametro)
#print(Sexto_Parametro)
#print(ERROR)
#print(Matriz_resultados)
````
